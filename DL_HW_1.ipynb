```python
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = "image.jpeg"
original_image = Image.open(image_path)

# Аугментации
augmentation_pipeline = transforms.Compose([
    transforms.Resize((256,256)),
    transforms.RandomRotation(degrees=30),
    transforms.RandomPerspective(distortion_scale=0.292, p=0.5),
    transforms.ColorJitter(brightness=0.444, contrast=0.8, saturation=0.533, hue=0.1),
    transforms.GaussianBlur(kernel_size=3, sigma=(0.179, 2.28)),
])

augmented_image = augmentation_pipeline(original_image)

# Визуализация
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Оригинальное изображение")
plt.imshow(np.array(original_image))
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title("Аугментированное изображение")
plt.imshow(np.array(augmented_image))
plt.axis('off')
plt.show()
```

Аугментация изображений значительно увеличивает вариантивность, а различные преобразования позволяют нам стать менее чувствительными к определенным факторам, важным, в зависимости от задачи, но при этом нужно использовать аугментацию аккуратно, не искажая изображения слишком сильно (как на нашем примере). Также можно упомянуть то, что некоторые аугментации позволяют регуляризировать изображения (например блюр), что позволяет модели не переобучаться из-за мелких деталей.
